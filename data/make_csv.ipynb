```python
import sqlite3
import csv
```

```python
# スクレイピングで取得したデータのリスト
scraping_data_list = []

# ローカル（自分のMac）
path = '/Users/kotomi/Unv/Lecture/ds-programming Q4/final/ds-programming_final'

# DBファイル名
db_name = 'scraping_data.sqlite'


# 1．DBに接続する
con = sqlite3.connect(path + db_name)

# 2．SQLを実行するためのオブジェクトを取得
cur = con.cursor()

# 3．SQLを用意
sql_select = 'SELECT * FROM scraping;'

# 4．SQLを実行
cur.execute(sql_select)

for r in cur:
  scraping_data_list.append(r)
  print(r)

# 6．DBへの接続を閉じる
con.close()
```

```python
with open('/Users/kotomi/Unv/Lecture/ds-programming Q4/final/ds-programming_final/data/scraping_data.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['日付', '現地平均気圧', '海面平均気圧', '合計降水量', '最大降水量', '最小降水量', '平均気温', '最高気温', '最低気温', '平均湿度', '最小湿度', '平均風速', '最大風速', '最大風速(風向)', '最大瞬間風速', '最大瞬間風速(風向)', '日照時間(h)', '降雪量(cm)', '最新積雪値(cm)', '天気概要昼', '天気概要夜'])
    writer.writerows(scraping_data_list)
```
