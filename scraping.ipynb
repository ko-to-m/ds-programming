```python
from bs4 import BeautifulSoup
import requests # HTTP操作用
import time
```

```python
# アクセスしたいWebサイトのURL
url = 'https://www.data.jma.go.jp/obd/stats/etrn/view/daily_s1.php?prec_no=46&block_no=47670&year=2024&month=1&day=&view='

# Webサーバにリクエストを出す．レスポンスを変数に格納しておく
r = requests.get(url)

# 文字化け防止
r.encoding = r.apparent_encoding
```

```python
print(f'レスポンスのコンテンツ:{r.text}')
```

DBを作成

```python
# 1．DBに接続する
con = sqlite3.connect(path + db_name)
# print(type(con))

# 2．SQLを実行するためのオブジェクトを取得
cur = con.cursor()

# 3．実行したいSQLを用意する
# テーブルを作成するSQL
# CREATE TABLE テーブル名（カラム名 型，...）;
sql_create_table_scraping = 'CREATE TABLE scraping(日付 int, 現地平均気圧 real, 海面平均気圧 real, 合計降水量 real, 最大降水量 real, 最小降水量 real, 平均気温 real, 最高気温 real, 最低気温 real, 平均湿度 int, 最小湿度 int, 平均風速 real, 最大風速 real, 最大風速風向 text, 最大瞬間風速 real, 最大瞬間風速風向 text, 日照時間h real, 降雪量cm int, 最新積雪値cm int, 天気概要昼 text, 天気概要夜 text);'

# 4．SQLを実行する
cur.execute(sql_create_table_scraping)

# 5．必要があればコミットする（データ変更等があった場合）
# 今回は必要なし

# 6．DBへの接続を閉じる
con.close()
```

DBにデータを挿入

```python
# 1.DBに接続する
con = sqlite3.connect(path + db_name)
print(type(con))

# 2．SQLを実行するためのオブジェクトを取得
cur = con.cursor()

# 3.SQLを用意
# データを挿入するSQL
# INSERT INTO テーブル名 VALUES(列に対応したデータをカンマ区切りで);
sql_insert_many = "INSERT INTO scraping VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);"

# 用意したリスト
value_new_list

# 4.SQLを実行
cur.executemany(sql_insert_many, value_new_list)

# 5.コミット処理(データ操作を反映させる)
con.commit()

# 6.DBへの接続を閉じる
con.close()
```

DB内のデータを参照

```python
# 1．DBに接続する
con = sqlite3.connect(path + db_name)

# 2．SQLを実行するためのオブジェクトを取得
cur = con.cursor()

# 3．SQLを用意
sql_select = 'SELECT * FROM scraping;'

# 4．SQLを実行
cur.execute(sql_select)

for r in cur:
  print(r)

# 6．DBへの接続を閉じる
con.close()
```
