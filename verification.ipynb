```python
import pandas as pd
from matplotlib import pyplot as plt
from random import randint
```

```python
my_data = pd.read_csv('/Users/kotomi/Unv/Lecture/ds-programming Q4/final/ds-programming_final/data/my_data.csv')
my_data
```

```python
# 平均気温、最高気温、最低気温を追加
my_data['Average'] = my_data['気温']
my_data['Max'] = my_data['気温']
my_data['min'] = my_data['気温']
```

```python
my_data
```

```python
# 複数値があるものの平均気温を出す
# 13日
a = 0
for i in range(len(my_data.iloc[5:9, 1])):
    a += my_data.iloc[5 + i, 1]

average_13 = round(a/len(my_data.iloc[5:9, 1]), 1)
print(average_13)


# 14日
b = 0
for i in range(len(my_data.iloc[9:12, 1])):
    b += my_data.iloc[9 + i, 1]

average_14 = round(b/len(my_data.iloc[9:12, 1]), 1)
print(average_14)

# 15日
c = 0
for i in range(len(my_data.iloc[12:14, 1])):
    c += my_data.iloc[12 + i, 1]

average_15 = round(c/len(my_data.iloc[12:14, 1]), 1)
print(average_15)
```

```python
# 平均気温、最高気温、最低気温を追加
# 13日
my_data.iloc[5:9, 2] = average_13
my_data.iloc[5:9, 3] = round(max(my_data.iloc[5:9, 1]), 1)
my_data.iloc[5:9, 4] = round(min(my_data.iloc[5:9, 1]), 1)

# 14日
my_data.iloc[9:12, 2] = average_14
my_data.iloc[9:12, 3] = round(max(my_data.iloc[9:12, 1]), 1)
my_data.iloc[9:12, 4] = round(min(my_data.iloc[9:12, 1]), 1)

# 15日
my_data.iloc[12:14, 2] = average_15
my_data.iloc[12:14, 3] = round(max(my_data.iloc[12:14, 1]), 1)
my_data.iloc[12:14, 4] = round(min(my_data.iloc[12:14, 1]), 1)
```

```python
n_my_data = my_data.drop('気温', axis=1)
```

```python
new_my_data = n_my_data.drop([6, 7, 8, 10, 11, 13, 12])
```

```python
new_my_data = new_my_data.rename(columns={'日付': 'date'})
```

```python
scraping_data = pd.read_csv('/Users/kotomi/Unv/Lecture/ds-programming Q4/final/ds-programming_final/data/scraping_data.csv')
scraping_data
```

```python
new_scraping_data = scraping_data[['日付', '平均気温','最高気温', '最低気温']]
new_scraping_data = new_scraping_data.rename(columns={'日付': 'date', '平均気温': 'Average', '最高気温': 'Max', '最低気温': 'min'})
new_scraping_data
```

```python
new_my_data
```

```python
# 平均値の折れ線グラフ
x1 = new_scraping_data['date']
y1 = new_scraping_data['Average']
x2 = new_my_data['date']
y2 = new_my_data['Average']

plt.title("Average")
plt.plot(x1, y1, label="scraping_data")
plt.plot(x2, y2, label="my_data")
plt.legend()
```

```python
# scrapingしたデータとmy_dataの数を同じにするためnew_scraping_dataを変える
new_scraping_data_ = new_scraping_data[(new_scraping_data['date'] == 5) | (new_scraping_data['date'] == 6) | (new_scraping_data['date'] == 8) | (new_scraping_data['date'] == 10) | 
                  (new_scraping_data['date'] == 11) | (new_scraping_data['date'] == 13) | (new_scraping_data['date'] == 14)]
```

```python
print('new_scraping_data_')
print(f'{new_scraping_data_}\n')
print('new_my_data')
print(new_my_data)
```

```python
def corr(a, b): # 散布図と相関係数を出す関数
    # 散布図
    x = new_scraping_data_[a]
    y = new_my_data[a]

    plt.title(a)
    plt.xlabel("scraping_data")
    plt.ylabel("my_data")
    plt.scatter(x, y)


    # 相関係数
    res = x.corr(y)
    print(f'{b}気温の相関係数：{res}')
```

```python
# 平均気温
corr('Average', '平均')
```

```python
# 最高気温
corr('Max', '最高')
```

```python
# 最低
corr('min', '最低')
```
